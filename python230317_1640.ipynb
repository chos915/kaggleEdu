## 데이터 핸들링
- 행과 열을 다루는 방법

```python
lemonade.head()
```

- 새로운 칼럼 추가

```python
lemonade['Sold'] = 0
lemonade.head()
```

```python
lemonade['Sold'] = lemonade['Lemon'] + lemonade['Orange']
lemonade.head(3)
```

- Revenue 칼럼 : 매출 계산 

```python
lemonade['Revenue'] = lemonade['Price'] * lemonade['Sold']
lemonade.head(3)
```

```python
lemonade.head()
```

```python
        #               행 추출 조건식                열 추출
result = lemonade.loc[lemonade['Revenue'] > 100, ['Date', 'Revenue']].reset_index(drop=True)
result
```

```python
cols = [ 'Temperature', 'Date', 'Revenue',]
result = lemonade.loc[lemonade['Revenue'] > 100, cols].reset_index(drop=True)
result
```

### 집계함수(Aggregation)
- 그룹바이 및 집계 --> 피벗테이블

```python
import numpy as np 
lemonade.groupby('Location')['Revenue'].agg([max, min, np.mean])
```

```python
import numpy as np 
lemonade.groupby('Location')[['Revenue', 'Temperature']].agg([max, min, np.mean, np.std, sum])
```
